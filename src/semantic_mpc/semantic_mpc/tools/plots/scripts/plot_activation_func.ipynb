```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha_s = 10
d_th = 9.0

# Range of values
d_min = np.linspace(0, 10, 400)
d_min_q = np.linspace(0, 10, 400)

# Functions
Q_d = 1e-2 / (1 + np.exp(alpha_s * (-d_min**2 + d_th)))
Q_d_q = 1 / (1 + np.exp(alpha_s * (-d_min_q**2 + d_th)))
C_a = Q_d * d_min**2

# Create figure
fig, axs = plt.subplots(1, 2, figsize=(4.5, 1.5))

# Plot Q_d
axs[0].plot(d_min_q, Q_d_q, color="royalblue", linewidth=2.5)
axs[0].set_xlabel(r'$d_{\min}$')
axs[0].set_ylabel(r'$s(\cdot)$')
axs[0].set_xlim(0, 10)
axs[0].grid(True, linewidth=0.1)

# Plot C_a
axs[1].plot(d_min, C_a, color='orange', linewidth=2.5)
axs[1].set_xlabel(r'$d_{\min}$')
axs[1].set_ylabel(r'$J_a(\cdot)$')
axs[1].set_xlim(0, 10)
axs[1].grid(True, linewidth=0.1)

plt.tight_layout()
output_path = "activation_function.svg"
plt.savefig(output_path, format='svg', dpi=1500, bbox_inches='tight')
plt.close()
```
